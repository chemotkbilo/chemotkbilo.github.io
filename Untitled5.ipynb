```python
'''
This code execute three different models and perform mean squared error comparison. The
data utilized is a time series of confirmed Covid-19 across multiple countries. The code
allows the selection of a country of interest and perform the following:
            - Linear Regression
            - Neural Network with 2 nodes
            - Neural Network with 4 nodes, 2 layers with 2 hidden nodes
Additionally, it creates a plot of time versus of prediction error for the following
            - A fit of the model on all of the selected country
            - A fit on the 1/2 of the country of interest, and then appkies to all of A

Author: GitHub - Chemotkbilo
'''

import pandas as pd
import numpy as np
import plotly.graph_objs as go
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_squared_error
from sklearn.linear_model import LinearRegression
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

url = "https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_confirmed_global.csv"
covid_df = pd.read_csv(url)

#selection the countries of interest, country A and country B
country_A = "US"
country_B = "Brazil"

#generating two dataframe for each country specifically to keep it clean
country_A_data = covid_df[covid_df["Country/Region"] == country_A]
country_B_data = covid_df[covid_df["Country/Region"] == country_B]
country_A_data = country_A_data.groupby("Country/Region").sum()
country_B_data = country_B_data.groupby("Country/Region").sum()

# defining a function to pre-process data such that it handles data formats such as cases and dates to be integers
def preprocess_data(country_covid):
    dates = country_covid.columns[3:]
    cases = country_covid.iloc[0, 3:].values.astype(int)
    return dates, cases

#invoking the preprocess data function on each country
dates_country_A, cases_country_A = preprocess_data(country_A_data)
dates_country_B, cases_country_B = preprocess_data(country_B_data)

#defining a function to handle the regression: Sources utilized are below
#https://scikit-learn.org/stable/tutorial/statistical_inference/supervised_learning.html
#https://machinelearningmastery.com/make-predictions-scikit-learn/
#https://stackoverflow.com/questions/65734567/what-is-python-sklearn-predict-function-working-principle
def fit_LR_model(x_train, y_train, x_test, model):
    model.fit(x_train, y_train)
    y_train = model.predict(x_train)
    y_test = model.predict(x_test)
    return y_train, y_test

#take the date and split them into test and train dataset using the following source
#https://scikit-learn.org/stable/modules/generated/sklearn.model_selection.train_test_split.html
x_country_A = np.arange(len(dates_country_A)).reshape(-1, 1)
y_country_A = cases_country_A.reshape(-1, 1)
x_country_A_train, x_country_A_test, y_country_A_train, y_country_A_test = train_test_split(x_country_A, y_country_A, test_size=0.5, shuffle=False)

#perform a standard data normalization:
#https://scikit-learn.org/stable/modules/generated/sklearn.preprocessing.StandardScaler.html
scaler = StandardScaler()
x_country_A_train_scaled = scaler.fit_transform(x_country_A_train)
x_country_A_test_scaled = scaler.transform(x_country_A_test)

#invoking the linear regression
LR = LinearRegression()
y_country_A_train_LR, y_country_A_test_LR = fit_LR_model(x_country_A_train_scaled, y_country_A_train, x_country_A_test_scaled, LR)

#creating a neural network function that allows specification for layers and nodes
# using the following link as the baseline https://faroit.com/keras-docs/1.0.1/getting-started/sequential-model-guide/
#https://scikit-learn.org/stable/modules/generated/sklearn.neural_network.MLPClassifier.html
#https://towardsdatascience.com/math-neural-network-from-scratch-in-python-d6da9f29ce65
#https://kavita-ganesan.com/neural-network-intro/
#https://www.activestate.com/resources/quick-reads/how-to-create-a-neural-network-in-python-with-and-without-keras/
def NN_model(nodes, layers):
    model = Sequential()
    #adding RELU activation
    model.add(Dense(nodes, activation='relu', input_shape=(1,)))
    #ensuring that the fully interconnected
    for _ in range(layers):
        model.add(Dense(nodes, activation='relu'))
     #specifying that it linear activation
    model.add(Dense(1, activation='linear'))
    #include mean squared error (MSE)
    model.compile(optimizer='adam', loss='mean_squared_error')
    return model

#neural network 2 nodes
NN_model_2_nodes = NN_model(2, 0)
#neural network 4 nodes over 2 layers with 2 hidden nodes
NN_model_4_nodes_2layers = NN_model(4, 2)

#invoking the function to perform the neural network modeling
#https://towardsdatascience.com/math-neural-network-from-scratch-in-python-d6da9f29ce65
#https://faroit.com/keras-docs/1.0.1/getting-started/sequential-model-guide/
NN_model_2_nodes.fit(x_country_A_train_scaled, y_country_A_train, epochs=100, batch_size=8, verbose=0)
NN_model_4_nodes_2layers.fit(x_country_A_train_scaled, y_country_A_train, epochs=100, batch_size=8, verbose=0)

#https://scikit-learn.org/stable/modules/generated/sklearn.preprocessing.StandardScaler.html
y_country_A_train_nn_2 = scaler.inverse_transform(NN_model_2_nodes.predict(x_country_A_train_scaled))
y_country_A_test_nn_2 = scaler.inverse_transform(NN_model_2_nodes.predict(x_country_A_test_scaled))

y_country_A_train_nn_4 = scaler.inverse_transform(NN_model_4_nodes_2layers.predict(x_country_A_train_scaled))
y_country_A_test_nn_4 = scaler.inverse_transform(NN_model_4_nodes_2layers.predict(x_country_A_test_scaled))


tr_p = [y_country_A_train_LR, y_country_A_train_nn_2, y_country_A_train_nn_4]
te_p = [y_country_A_test_LR, y_country_A_test_nn_2, y_country_A_test_nn_4]

mse = {
    'train': [],
    'test': []
}

#https://scikit-learn.org/stable/modules/generated/sklearn.metrics.mean_squared_error.html
for i, i in zip(tr_p, te_p):
    mse['train'].append(mean_squared_error(y_country_A_train, i))
    mse['test'].append(mean_squared_error(y_country_A_test, i))

mse_country_A = {
    'train_LR': mse['train'][0],
    'train_n_2': mse['train'][1],
    'train_n_4': mse['train'][2],
    'test_LR': mse['test'][0],
    'test_n_2': mse['test'][1],
    'test_n_4': mse['test'][2]
}

MSE_data = {
    "Model": ["Linear Regression (LR)", "NN (2 Nodes)", "NN(4 Nodes, 2 Layers)"],
    "Train MSE": [mse_country_A['train_LR'], mse_country_A['train_n_2'], mse_country_A['train_n_4']],
    "Test MSE": [mse_country_A['test_LR'], mse_country_A['test_n_2'], mse_country_A['test_n_4']]
}

```

```python
print(f"{'Model':<50}{'Train MSE':<20}{'Test MSE'}")
for i in range(len(MSE_data["Model"])):
    print("-----------------------------------------------------------------------------------------------------------------")
    print(f"{MSE_data['Model'][i]:<50}{format(MSE_data['Train MSE'][i], '.2e'):<20}{format(MSE_data['Test MSE'][i], '.2e')}")
    print("-----------------------------------------------------------------------------------------------------------------")
```

```python
import numpy as np
import plotly.graph_objects as go

def batch_plot(Trd, Ted, pd, mdl, cty):
    fig = go.Figure()
     #https://plotly.com/python/line-and-scatter/
    fig.add_trace(go.Scatter(x=np.arange(len(Trd)), y=Trd.flatten(), mode='lines', name='Train Data'))
    fig.add_trace(go.Scatter(x=np.arange(len(Trd), len(Trd) + len(Trd)), y=Ted.flatten(), mode='lines', name='Test Data'))

    for i, (Trd_pd, Ted_pd) in enumerate(pd):
        fig.add_trace(go.Scatter(x=np.arange(len(Trd)), y=Trd_pd.flatten(), mode='lines', name=f'{mdl[i]}: Train Prediction'))
        fig.add_trace(go.Scatter(x=np.arange(len(Trd), len(Trd) + len(Trd)), y=Ted_pd.flatten(), mode='lines', name=f'{mdl[i]}: Test Prediction'))

    # Update layout
    fig.update_layout(title=f"COVID-19 cases: Model predictions comparison for {cty}",
                      xaxis_title="Days",
                      yaxis_title="Number of Cases")

    # Show plot
    fig.show()

pd = [
    (y_country_A_train_LR, y_country_A_test_LR),
    (y_country_A_train_nn_2, y_country_A_test_nn_2),
    (y_country_A_train_nn_4, y_country_A_test_nn_4)
]
mdl = ['Linear Regression', 'Neural Network (2 Nodes)', 'Neural Network (4 Nodes, 2 Layers)']
cty = country_A

# Call the function to generate plots
batch_plot(y_country_A_train, y_country_A_test, pd, mdl, cty)
```
